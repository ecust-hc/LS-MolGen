```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Scaffolds import MurckoScaffold

suppl = Chem.SmilesMolSupplier('generate_10000.smi', titleLine=False)
mols = [m for m in suppl if m]
```

```python
basic_structure = mols[222]
atomic_scaffold = MurckoScaffold.GetScaffoldForMol(basic_structure)
atomic_scaffold.Compute2DCoords()
graph_scaffold = MurckoScaffold.MakeScaffoldGeneric(atomic_scaffold)
Draw.MolsToGridImage([basic_structure, atomic_scaffold, graph_scaffold])
```

```python
mols_atomic_scaffolds = [MurckoScaffold.GetScaffoldForMol(mol) for mol in mols]
for i in mols_atomic_scaffolds:
    i.Compute2DCoords()

def genericize_scaffold(s):
    try:
        return MurckoScaffold.MakeScaffoldGeneric(s)
    except ValueError:
        return None

mols_grafh_scaffolds = [genericize_scaffold(s) for s in mols_atomic_scaffolds]
print(len(mols), len(mols_atomic_scaffolds), len(mols_grafh_scaffolds), len([x for x in mols_grafh_scaffolds if x == None]))
Draw.MolsToGridImage([mols[111], mols_atomic_scaffolds[111], mols_grafh_scaffolds[111]])
```

```python
scaffold_smiles = [Chem.MolToSmiles(scaffold) for scaffold in mols_atomic_scaffolds if scaffold != None]
print(len(scaffold_smiles), scaffold_smiles[111])

import collections
counter=collections.Counter(scaffold_smiles)

print(counter)
```

```python
sort_counter = sorted(counter.items(),  key=lambda d: d[1], reverse=True)

key = [sort_counter[i][0] for i in range(len(sort_counter))]
value = [sort_counter[i][1] for i in range(len(sort_counter))]

print(len(key))
Draw.MolsToGridImage([Chem.MolFromSmiles(x) for x in key[:10]])
```

```python
import numpy as np  
import matplotlib.mlab as mlab  
import matplotlib.pyplot as plt  

fig, ax = plt.subplots(1, 1, figsize=(4, 3))

#ax.plot(range(10), value[:10], '.', ms=1)
plt.bar(range(10), value[:10], color='lightsteelblue')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel('Top 10 scaffolds', fontsize=14)
ax.set_ylabel('Count', fontsize=14)
ax.xaxis.set_ticklabels([])
ax.set_xticks([])


plt.tight_layout()
#fig.figure.savefig('scaffold_count.png', dpi=300)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
from matplotlib import style
import matplotlib.pyplot as plt
from rdkit import Chem, DataStructs
from rdkit.Chem import Descriptors
from rdkit.Chem import AllChem, Draw, PandasTools
from rdkit.Chem import Lipinski, MACCSkeys
from SA_Score import sascorer
from NP_Score import npscorer

df = pd.read_csv('generate_10000.smi', names=['SMILES'])

df['MW'] = df['SMILES'].apply(lambda x: Descriptors.MolWt(Chem.MolFromSmiles(x)))
df['MolLogP'] = df['SMILES'].apply(lambda x: Descriptors.MolLogP(Chem.MolFromSmiles(x)))
df['QED'] = df['SMILES'].apply(lambda x: Chem.QED.qed(Chem.MolFromSmiles(x)))
df['TPSA'] = df['SMILES'].apply(lambda x: Descriptors.TPSA(Chem.MolFromSmiles(x)))
df['SA'] = [sascorer.calculateScore(Chem.MolFromSmiles(m)) for m in df['SMILES']]
df['np-likeness'] = [npscorer.scoreMol(Chem.MolFromSmiles(m)) for m in df['SMILES']]
```

```python
fig, ax = plt.subplots(2, 3, figsize=(12, 8))
sns.kdeplot(df['MW'], ax=ax[0, 0], shade=True, label='mean= {:.2f}'.format(df['MW'].mean()))
ax[0, 0].legend()
sns.kdeplot(df['MolLogP'], ax=ax[0, 1], shade=True, label='mean= {:.2f}'.format(df['MolLogP'].mean()))
ax[0, 1].legend()
sns.kdeplot(df['QED'], ax=ax[0, 2], shade=True, label='mean= {:.2f}'.format(df['QED'].mean()))
ax[0, 2].legend()
sns.kdeplot(df['TPSA'], ax=ax[1, 0], shade=True, label='mean= {:.2f}'.format(df['TPSA'].mean()))
ax[1, 0].legend()
sns.kdeplot(df['SA'], ax=ax[1, 1], shade=True, label='mean= {:.2f}'.format(df['SA'].mean()))
ax[1, 1].legend()
sns.kdeplot(df['np-likeness'], ax=ax[1, 2], shade=True, label='mean= {:.2f}'.format(df['np-likeness'].mean()))
ax[1, 2].legend()

plt.tight_layout()
fig.figure.savefig('properties_10000_generated.png', dpi=300)
```
